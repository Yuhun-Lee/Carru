```python
import pandas as pd
from geopy.geocoders import Nominatim
import re

geo_local = Nominatim(user_agent='South Korea')

def extract_street_address(address):
    address = re.sub(r'(\w+구|\w+군|\w+면|\w+읍)\s*', '', address)
    match = re.search(r'.*(?:길|로)\s*\d+[-]?\d*', address)
    if match:
        return match.group()
    else:
        return None  # 오류가 있을 경우 None 반환
    
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0, 0]  # 오류 시 [0, 0] 반환

# cleaned_ware_info.xlsx 파일 읽기
cleaned_ware_info_df = pd.read_excel('Dataset/coupangData.xlsx')

# '소재지' column에 대해 geocoding 처리
cleaned_rows = []

print(f"전체 {len(cleaned_ware_info_df)}개의 주소에 대해 geocoding을 시작합니다.\n")

for idx, address in enumerate(cleaned_ware_info_df['소재지']):
    lat_lon = geocoding(extract_street_address(address))
    
    if lat_lon == [0, 0]:
        print(f"[{idx+1}/{len(cleaned_ware_info_df)}] 주소 '{address}' geocoding 실패 (위도, 경도: [0,0])")
        continue  # 위도, 경도 값이 [0,0]이면 해당 행을 건너뜀
    
    print(f"[{idx+1}/{len(cleaned_ware_info_df)}] 주소 '{address}' geocoding 성공 (위도: {lat_lon[0]}, 경도: {lat_lon[1]})")
    
    # 성공한 경우, 위도/경도 값 포함하여 행을 추가
    row = cleaned_ware_info_df.iloc[idx].copy()  # 원래 행을 복사하여 저장
    row['LAT'] = lat_lon[0]
    row['LONG'] = lat_lon[1]
    cleaned_rows.append(row)

# geocoding 성공한 행들로 새 DataFrame 생성
final_cleaned_df = pd.DataFrame(cleaned_rows)

print(f"\n처리 후 {len(final_cleaned_df)}개의 행이 남았습니다.\n")
print(final_cleaned_df)

# 결과를 새로운 엑셀 파일로 저장
final_cleaned_df.to_excel('Dataset/cleaned_coupangData.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 엑셀 파일에서 location과 ware_info 데이터 읽기
location_df = pd.read_excel('Dataset/cleaned_coupangData.xlsx')  # location
location_df = location_df[~location_df['소재지'].str.contains("제주")]

ware_info_df = pd.read_excel('Dataset/cleaned_ware_info_with_lat_long.xlsx')  # destination with lat/long
ware_info_df = ware_info_df[~ware_info_df['소재지'].str.contains("제주")]

item_type_df = pd.read_excel('Dataset/item_types.xlsx')  # item types
departure_info_df = pd.read_excel('Dataset/WAREHOUSE.xlsx')

locations = location_df['소재지'].values
names = item_type_df['취급품목'].values
today_date = datetime.now()

# 데이터 생성 함수
def create_logistics_data(num_entries):
    data = []
    product_id = 1
    
    for location, name in zip(locations, names):
        for i in range(num_entries):
            l_name = f"{name}_{i}"
            # 랜덤 destination 선택
            
            destination_row = ware_info_df.sample(n=1).iloc[0]
            destination = destination_row['소재지']
            destination_lat = destination_row['LAT']
            destination_long = destination_row['LONG']
                
            weight = np.random.randint(3, 10)  # 3톤에서 10톤 랜덤
    
            # 며칠 뒤의 날짜와 시간을 포함한 마감 기한 생성
            deadline_days = np.random.randint(1, 4)  # 1일에서 3일 랜덤
            deadline_time = today_date + timedelta(days=deadline_days, hours=np.random.randint(0, 24))
            
            # 출발지의 위도 및 경도 가져오기
            departure_row = departure_info_df[departure_info_df['location'] == location]
            if departure_row.empty:
                continue  # location이 없는 경우 생략
            departure_lat = departure_row['location_lat'].values[0]
            departure_long = departure_row['location_long'].values[0]

            # 데이터를 리스트에 추가
            data.append({
                'user_id': None,
                'product_id': product_id,
                'departure': location,
                'departure_lat': departure_lat,
                'departure_long': departure_long,
                'name': l_name,
                'destination': destination,
                'destination_lat': destination_lat,
                'destination_long': destination_long,
                'weight': weight,
                'deadline': deadline_time.strftime('%Y-%m-%d %H:00'),
                'date': today_date.strftime('%Y-%m-%d')
            })
            
            product_id += 1

    # DataFrame으로 반환
    return pd.DataFrame(data)

def assign_user_ids(df, entries_per_user=100):
    df = df.copy()
    df['user_id'] = (df.index // entries_per_user) + 1
    return df

logistics_data = create_logistics_data(100)
logistics_data_with_ids = assign_user_ids(logistics_data)

logistics_data_with_ids.to_excel('Dataset/PRODUCT.xlsx', index=False)
```

```python
import pandas as pd
from geopy.distance import geodesic

# 거리 계산 함수
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).km

# 무게와 거리를 고려한 단가 계산 함수
def calculate_shipping_cost(weight_class, distance):
    # 가격 테이블 (단위: 만 원)
    price_table = [
        [4, 5, 7, 8, 11, 13, 18, 20],  # 20km
        [5, 6, 8, 9, 12, 14, 19, 21],  # 30km
        [6, 7, 9, 10, 13, 15, 20, 22], # 50km
        [7, 8, 10, 11, 14, 17, 21, 23], # 70km
        [8, 9, 11, 12, 16, 19, 22, 25], # 90km
        [9, 10, 12, 13, 17, 20, 24, 27], # 110km
        [10, 11, 13, 14, 18, 21, 25, 28], # 130km
        [11, 12, 14, 15, 19, 23, 27, 30], # 150km
        [12, 13, 15, 16, 20, 25, 28, 31], # 170km
        [13, 14, 18, 19, 23, 27, 30, 32], # 200km
        [15, 16, 19, 20, 25, 33, 39, 42], # 250km
        [17, 18, 25, 26, 30, 37, 45, 48], # 300km
        [19, 20, 27, 28, 33, 38, 50, 53], # 350km
        [20, 21, 28, 30, 35, 42, 55, 58], # 400km
        [22, 23, 30, 32, 37, 44, 58, 62], # 450km
        [25, 26, 32, 34, 38, 46, 60, 64], # 500km
        [27, 28, 35, 37, 41, 48, 63, 67], # 550km
        [30, 31, 38, 40, 43, 50, 65, 70]  # 600km
    ]
    
    # 거리와 화물 무게에 대한 인덱스 계산
    distance_index = {
        20: 0, 30: 1, 50: 2, 70: 3, 90: 4,
        110: 5, 130: 6, 150: 7, 170: 8,
        200: 9, 250: 10, 300: 11, 350: 12,
        400: 13, 450: 14, 500: 15, 550: 16, 600: 17
    }
    
    # 화물 무게 클래스를 인덱스로 변환
    weight_index = {
        1: 0, 1.4: 1, 2.5: 2, 3.5: 3, 5: 4, 11: 5, 18: 6, 25: 7
    }
    
    # 거리와 화물 무게의 유효한 인덱스 찾기
    if distance not in distance_index:
        next_distance = min([d for d in distance_index.keys() if d > distance], default=None)
        if next_distance is not None:
            distance = next_distance
    
    if weight_class not in weight_index:
        next_weight = min([w for w in weight_index.keys() if w > weight_class], default=None)
        if next_weight is not None:
            weight_class = next_weight
            
    # 인덱스 얻기
    distance_index_value = distance_index.get(distance, len(price_table) - 1)
    weight_index_value = weight_index.get(weight_class, len(price_table[0]) - 1)
    
    # 가격 계산
    price = price_table[distance_index_value][weight_index_value]
    
    return price

# 데이터 읽기
product_file_path = "Dataset/PRODUCT.xlsx"
product_df = pd.read_excel(product_file_path)

product_df['distance'] = product_df.apply(
    lambda row: calculate_distance(row['departure_lat'], row['departure_long'],
                                   row['destination_lat'], row['destination_long']), axis=1)

# 운송 비용 계산 후 새로운 열 추가
product_df['price'] = product_df.apply(
    lambda row: calculate_shipping_cost(row['weight'], row['distance']), axis=1)

# 결과 저장
output_file_path = "PRODUCT.xlsx"
product_df.to_excel(output_file_path, index=False)
```

```python
import random
import string
from korean_name_generator import namer

def generate_random_name(s):  # 1이면 남자 이름, 0이면 여자 이름 
    if s == 1:  # 남자 이름
        name = namer.generate(True)
    else:  # 여자 이름 
        name = namer.generate(False)
    
    return name 

def generate_random_email():
    # 이메일 아이디의 길이 설정 (7자 이상 12자 이하)
    email_id_length = random.randint(7, 12)
    # 이메일 아이디 생성
    email_id = ''.join(random.choices(string.ascii_lowercase + string.digits, k=email_id_length))
    # 도메인 선택
    domain = random.choice(["@naver.com", "@gmail.com", "@daum.net"])
    # 완성된 이메일 반환
    return email_id + domain

def generate_random_phone_number():
    # 4자리 숫자 두 개 생성
    middle_part = random.randint(1000, 9999)
    last_part = random.randint(1000, 9999)
    # 완성된 전화번호 반환
    return f"010-{middle_part}-{last_part}"
```

```python
# 2. 화주 데이터셋 생성: 창고 주소 랜덤 매칭 - many to one 

import numpy as np
import pandas as pd

ware_info_df = pd.read_excel('Dataset/WAREHOUSE.xlsx') 

def create_random_dataset(num_rows):
    selected_rows = ware_info_df.sample(n=num_rows, replace=True).reset_index(drop=True)
    
    data = {
        "NAME": [generate_random_name(1) for _ in range(num_rows)],
        "EMAIL": [generate_random_email() for _ in range(num_rows)],
        "PHONE_NUMBER": [generate_random_phone_number() for _ in range(num_rows)],
        "USER_STATUS": [0 for _ in range(num_rows)],    # 화주
        "LOCATION": selected_rows['location'],
        "LOCATION_LAT": selected_rows['location_lat'],
        "LOCATION_LONG": selected_rows['location_long']
    }
    return pd.DataFrame(data)

df = create_random_dataset(300)
df.to_excel('Dataset/USER_Owner_mto1.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np

# 파일 로드
user_df = pd.read_excel('Dataset/USER.xlsx')         # 사용자 정보 파일
ware_info_df = pd.read_excel('Dataset/cleaned_ware_info_with_lat_long.xlsx') # 창고 정보 파일
ware_info_df = ware_info_df[~ware_info_df['소재지'].str.contains("제주")]
garage_df = pd.read_excel('Dataset/garage.xls')       # 차고 정보 파일
garage_df = garage_df[~garage_df['주소'].str.contains("제주")]

def create_input_dataset(num_rows):
    driver_user_ids = user_df[user_df['USER_STATUS'] == 1]['USER_ID']
    selected_departure = ware_info_df.sample(n=num_rows, replace=True).reset_index(drop=True)
    selected_destination = garage_df.sample(n=num_rows, replace=True).reset_index(drop=True)
    
    # 데이터 생성
    data = {
        "user_id": np.random.choice(driver_user_ids, num_rows),  
        "max_weight": np.random.randint(10, 25, num_rows),  # 최대 수용 무게 (20~25)
        "min_weight": np.random.randint(2, 10, num_rows),   # 최소 수용 무게 (2~5)
        "departure": selected_departure['소재지'],                    # ware_info에서 무작위 선택된 출발지 주소
        "departure_lat": selected_departure['LAT'],                  # 출발지 위도
        "departure_long": selected_departure['LONG'],                # 출발지 경도
        "destination": selected_destination['주소'],                  # garage에서 무작위 선택된 목적지 주소
        "destination_lat": selected_destination['LAT'],              # 목적지 위도
        "destination_long": selected_destination['LONG'],            # 목적지 경도
        "preference": np.random.randint(0, 11, num_rows)             # 선호도 (0~10)
    }
    
    return pd.DataFrame(data)

# INPUT 데이터셋 생성 및 엑셀 파일로 저장
num_rows = 10000  # 생성할 데이터 수
input_df = create_input_dataset(num_rows)
input_df.to_excel('INPUT.xlsx', index=False)

print("INPUT.xlsx 파일이 성공적으로 생성되었습니다.")
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

import re

def extract_street_address(address):
    address = re.sub(r'(\w+구|\w+군|\w+면|\w+읍)\s*', '', address)
    match = re.search(r'.*(?:길|로)\s*\d+[-]?\d*', address)
    if match:
        return match.group()
    else:
        return None  # 오류가 있을 경우 None 반환

    
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except:
        return [0, 0]  # 오류 시 [0, 0] 반환

address = "오태1길 48"
result = extract_street_address(address)
print(extract_street_address(address))
print(geocoding(result))
```
